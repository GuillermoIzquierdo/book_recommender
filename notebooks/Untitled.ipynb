```python
## Imports

# Basics
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_comics = pd.read_csv('clean_comics_reviews.csv')
df_fantasy = pd.read_csv('clean_fantasy_paranormal_reviews.csv')
df_history = pd.read_csv('clean_history_bio_reviews.csv')
df_mistery = pd.read_csv('clean_mistery_thriller_crime_reviews.csv')
df_poetry = pd.read_csv('clean_poetry_reviews.csv')
df_romance = pd.read_csv('clean_romance_reviews.csv')
df_young = pd.read_csv('clean_young_adult_reviews.csv')
df_books = pd.read_csv('books_info_books_cleaned.csv')
```

```python
df_books.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'], inplace = True)
df_fantasy.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'], inplace = True)
df_history.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'], inplace = True)
df_mistery.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'], inplace = True)
df_poetry.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'], inplace = True)
df_romance.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'], inplace = True)
df_young.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'], inplace = True)
```

```python
df_fantasy['genre'] = 'fantasy'
df_history['genre'] = 'hisotry'
df_mistery['genre'] = 'mistery'
df_poetry['genre'] = 'poetry'
df_romance['genre'] = 'romance'
df_young['genre'] = 'young'
```

```python
merge = df_books.merge(
    pd.concat([df_fantasy, df_history, df_mistery, df_poetry, df_romance, df_young]), on = 'book_id'
)
```

```python
book_id_more_than_ten = merge.groupby('book_id').count()[
    merge.groupby('book_id').count()['review_id'] > 10
].index.astype('object')
```

```python
df_reviews = merge.set_index('book_id').loc[book_id_more_than_ten].reset_index()
df_reviews.head()
```

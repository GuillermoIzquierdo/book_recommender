```python
def fetch_book(isbn):
    url = "https://openlibrary.org/api/books"
    params = {
        'bibkeys': f"ISBN:{isbn}",
        'format': 'json',
        'jscmd': 'data'
    }
    response = requests.get(url, params=params).json()
    if f"ISBN:{isbn}" in response:
        return response[f"ISBN:{isbn}"]
    else:
        return None
```

```python
def fetch_book(isbn):
    url = "https://openlibrary.org/api/books"
    params = {
        'bibkeys': f"ISBN:{isbn}",
        'format': 'json',
        'jscmd': 'data'
    }
    response = requests.get(url, params=params).json()
    if f"ISBN:{isbn}" in response:
        return response[f"ISBN:{isbn}"]
    else:
        return ''
```

```python
%%time

# TODO: for row in rows => fetch_book => add column

for index, row in df_books.iterrows():
    if row["cover_url"] is None:
        isbn = row['isbn']
        print(f"Fetching cover for {row['title']}")
        book = fetch_book(isbn)
        if book:
            cover_url = book.get("cover", {}).get("large", "")
            df_books.loc[index, "cover_url"] = cover_url
        else:
            df_books.loc[index, "cover_url"] = ""
```

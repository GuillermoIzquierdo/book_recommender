```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.goodreads.com/list/show/264.Books_That_Everyone_Should_Read_At_Least_Once')
```

```python
soup = BeautifulSoup(response.content)
```

```python
soup.find_all('a', class_ = 'bookTitle')
```

```python
project_href = [i['href'] for i in soup.find_all('a', class_='bookTitle', href=True)]
print(project_href)
```
